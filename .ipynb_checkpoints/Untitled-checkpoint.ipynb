```python
import json

class Analyscratch:
    def __init__(self, filename):
        self.filename = filename
        try:
            with open(self.filename) as json_file:
                json_data = json.load(json_file)
            self.data_targets = json_data['targets']
        except IOError:
            print('An error occured trying to read the file.')

    def generic(self, data_name):
        self.info = {}
        self.actors = []
        
        if(data_name == "blocks"):
            self.data = {}
            for actor in self.data_targets:
                actor_name = actor['name']
                self.actors.append(actor)
                for data in actor[data_name]:
                    element = actor[data_name][data]
                    self.data[data] = element
                self.info[actor_name] = self.data
                self.data = {}
            return self.info
        
        elif(data_name == "costumes" or data_name == "sounds"):
            self.data = []
            for actor in self.data_targets:
                actor_name = actor['name']
                self.actors.append(actor)
                for data in actor[data_name]:
                    self.data.append(data)
                self.info[actor_name] = self.data
                self.data = []
            return self.info
        
        elif(data_name == "variables"):
            self.data = {}
            for actor in self.data_targets:
                actor_name = actor['name']
                self.actors.append(actor)
                for data in actor[data_name]:
                    element = actor[data_name][data]
                    self.data[data] = element
                self.info[actor_name] = self.data
                self.data = {}
            return self.info
        
        elif(data_name == "lists"):
            self.data = {}
            for actor in self.data_targets:
                actor_name = actor['name']
                self.actors.append(actor)
                for data in actor[data_name]:
                    element = actor[data_name][data]
                    self.data[data] = element
                self.info[actor_name] = self.data
                self.data = {}
            return self.info
        
        elif(data_name == "broadcasts"):
            self.data = {}
            for actor in self.data_targets:
                actor_name = actor['name']
                self.actors.append(actor)
                for data in actor[data_name]:
                    element = actor[data_name][data]
                    self.data[data] = element
                self.info[actor_name] = self.data
                self.data = {}
            return self.info
        
        elif(data_name == "comments"):
            self.data = {}
            for actor in self.data_targets:
                actor_name = actor['name']
                self.actors.append(actor)
                for data in actor[data_name]:
                    element = actor[data_name][data]
                    self.data[data] = element
                self.info[actor_name] = self.data
                self.data = {}
            return self.info
            
    def blocks(self, actorName = None, data = 'opcode'):
        self.info = self.generic("blocks")
        self.blocks = []
        self.data = {}
        
        for actor in self.info:
            elements = self.info[actor]
            
            if(data == 'opcode'):
                if(actorName):
                    if(actor == actorName):
                        for blocks in elements:
                            if type(elements[blocks]) is dict:
                                block = elements[blocks][data]
                                self.blocks.append(block)
                            else:
                                block = elements[blocks]
                                self.blocks.append('variable_' + block[1])

                else:
                    for blocks in elements:
                        if type(elements[blocks]) is dict:
                            block = elements[blocks][data]
                            self.blocks.append(block)
                        else:
                            block = elements[blocks]
                            self.blocks.append('variable_' + block[1])
                return self.blocks
            
            elif(data == 'next'):
                self.blocks = {}
                if(actorName):
                    if(actor == actorName):
                        for blocks in elements:
                            if type(elements[blocks]) is dict:
                                block = elements[blocks][data]
                                self.blocks[blocks] = block
                            else:
                                block = elements[blocks]
                                self.blocks[blocks] = 'None'
 ## CONTINUES FROM HERE - TODO
                else:
                    for blocks in elements:
                        if type(elements[blocks]) is dict:
                            block = elements[blocks][data]
                            self.blocks.append(block)
                        else:
                            block = elements[blocks]
                            self.blocks.append('variable_' + block[1])
                return self.blocks

    def listActors(self):
        self.actors = []
        for actor in self.data_targets:
            self.actors.append(actor['name'])
        return self.actors
    
    def listCostumes(self, actorName = None):
        self.costumes = []
        for actor in self.data_targets:
            if(actorName):
                if(actor['name'] == actorName):
                    for costumes in actor['costumes']:
                        self.costumes.append(costumes['name'])
            else:
                for costumes in actor['costumes']:
                    self.costumes.append(costumes['name'])
        return self.costumes
    
    def listSounds(self, actorName = None):
        self.sounds = []
        for actor in self.data_targets:
            if(actorName):
                if(actor['name'] == actorName):
                    for sounds in actor['sounds']:
                        self.sounds.append(sounds['name'])
    
    

block_list = Analyscratch('project.json')
block_list.blocks('Stage', 'next')
```
